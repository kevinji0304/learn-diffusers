Learn to use pipelines: pretrained model, select steps, and generate the image.

```python
from diffusers import DDPMPipeline
from IPython.display import display
from PIL import Image

ddpm = DDPMPipeline.from_pretrained('google/ddpm-cat-256')

for i in range(1, 25):
    print(i)
    image = ddpm(num_inference_steps=i).images[0]
    image.save(f"step_{i}.png")
    display(image)
```
