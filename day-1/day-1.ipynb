From pre-trained model `google/ddpm-cat-256`, we adjust the interferance steps and see what the diffusion model generate images with different noise levels.

```python
from diffusers import DDPMPipeline
from IPython.display import display
from PIL import Image

ddpm = DDPMPipeline.from_pretrained('google/ddpm-cat-256')

# for i in range(1, 25):
#     print(i)
#     image = ddpm(num_inference_steps=i).images[0]
#     image.save(f"step_{i}.png")
#     display(image)

image = ddpm(num_inference_steps=75).images[0]
image
```

By observing the generated images, we can see that the inteference steps are related to the noise level. More steps lead to clearer images, but not necessarily more recognizable.
